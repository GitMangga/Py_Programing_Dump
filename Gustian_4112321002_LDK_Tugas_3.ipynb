```python
!pip install newspaper3k
from newspaper import Article
```

```python
#Menambahkan tautan/link berita untuk di scraping
article = Article('https://wartakota.tribunnews.com/2022/09/15/erigo-x-tampilkan-balutan-proses-metamorfosis-di-new-york-fashion-week-the-show-spring', 'id')

#Melakukan parsing
article.download()
article.parse()

#Mengambil data untuk wordcloud 
berita = article.text
print('Teks berita:', berita)
```

```python
!pip install wordcloud==1.8.0
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from matplotlib import pyplot as plt
wc = WordCloud(max_font_size=1000, max_words=10000, background_color="white")
wc.generate(berita)
```

```python
plt.figure(figsize = (15, 15))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
